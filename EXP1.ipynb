```python
"""Expected Input/Output:
Case 1: When the stack is empty, check for stack underflow
Case 2: When the stack is full, check for stack overflow
Case 3: Print intermediate content of the stack after doing some push and pop
operations."""
```

```python
class Stack:
    def __init__(self, capacity):
        self.capacity = capacity
        self.stack = [None] * capacity
        self.top = -1

    def is_full(self):
        return self.top == self.capacity - 1

    def is_empty(self):
        return self.top == -1

    def push(self, value):
        if self.is_full():
            print(f"Error: Overflow! Cannot push {value}. Stack is full.")
        else:
            self.top += 1
            self.stack[self.top] = value
            print(f"Pushed: {value}")

    def pop(self):
        if self.is_empty():
            print("Error: Underflow! Stack is empty. Cannot pop.")
            return None
        else:
            x = self.stack[self.top]
            self.stack[self.top] = None # Clear reference
            self.top -= 1
            print(f"Popped: {x}")
            return x

    def display(self):
        if self.is_empty():
            print("Stack is Empty.")
        else:
            print("Stack elements (Top to Bottom):")
            for i in range(self.top, -1, -1):
                print(self.stack[i])
        print("---") # Separator


# Create a stack with a small capacity for easy demonstration
my_stack = Stack(3)
print("Jinto Joseph \nURK24CS1210")

print("--- Case 1: When the stack is empty, check for stack underflow ---")
my_stack.display() # Display empty stack
my_stack.pop()     # Attempt to pop from empty stack
print("\n")

print("--- Case 3: Print intermediate content of the stack after doing some push and pop operations ---")
my_stack.push(10)
my_stack.push(20)
my_stack.display() # Display after two pushes

my_stack.pop()     # Pop one element
my_stack.display() # Display after one pop

my_stack.push(30)
my_stack.push(40)  # This will fill the stack
my_stack.display()

print("\n--- Case 2: When the stack is full, check for stack overflow ---")
my_stack.push(50)  # Attempt to push to a full stack
my_stack.display() # Show stack content after attempted overflow push
print("\n")

print("--- Final state after some more pops ---")
my_stack.pop()
my_stack.pop()
my_stack.display() # Should have one element left
my_stack.pop()
my_stack.display() # Should be empty again
my_stack.pop() # Attempt underflow again
```
